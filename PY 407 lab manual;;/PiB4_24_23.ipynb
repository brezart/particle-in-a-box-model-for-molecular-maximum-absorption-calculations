# Background

### "A quantum review of molecular bonding"

Carbon based molecules tend to form covalent bonds which can be divided into two methods of
bonding, Sigma and Pi-bonds. Sigma-bonds are the most stable form of covalent bonding and are
the first bond formed between two carbons. This means that all single bonds in our
molecules are considered Sigma-bonds. Sigma-bonds are strong because their spatial
wavefunction overlap and their probability density is relatively high compared to Pi-bonds. This forms
strong bonds but leaves the bonded electrons (Sigma-electrons) localized. These Sigma-bonds
form the "backbone" of the molecule and give the molecule its structure. On the other hand, Pi-
bonds are formed above and below the "backbone" and the Pi-electron's spatial
probability density is highest above or below the Sigma-bonds. In our molecules, the other
bonds formed are called Pi-bonds and this is crucial for being able to apply our PiB model to
molecules. When carbon forms chains and rings like in our molecules, they begin to form what is
known as "Pi-conjugation." This means that the Pi-electrons are able to move between these
successive Pi-bonds along the backbone of the molecule to form delocalized wavefunctions
across the carbon bonds. This is where the PiB model begins to apply to our molecules. The
delocalized Pi-electrons are able to move freely in our molecular "box" but are still bound by the
Sigma bonds’ backbone.

These Sigma and Pi-bonds represent different energies that can be found within the system.
Since Sigma-bonds are stronger and formed first, they form the lower most electronic states.
Whereas Pi-bonds are the states closer to the highest occupied molecular
orbital (HOMO) and are the first to become excited (through absorption of a photon)
into the lowest unoccupied molecular orbit (LUMO). These
Pi-electrons are responsible for photon absorption and emission within the visible portion of the
electro-magnetic (E-M) spectrum and help to define the spectroscopic shape of the molecules.

## “Spectroscopy 101”

Wikipedia describes spectroscopy as the field of study that measures and interprets the electro-magnetic spectra that result from the interaction between electro-magnetic radiation and matter
as a function of the wavelength or frequency of the radiation (https://en.wikipedia.org/wiki/Spectroscopy). This is an
important tool for learning more about the structural and electronic information of a material
through its spectral data. For instance, we define “white” light as a superposition of
electromagnetic waves in the visible region (400-800nm).
As wavelengths within this visible region are absorbed/transmitted or reflected, we can see
different colors with our eyes. An interesting example of this phenomenon can be found in
plants. Plants contain chlorophyll a and b which absorb much of the blue and red wavelengths.
This allows the greens and yellows to be reflected back which our eyes perceive as the plant’s
color.

## Tying it all together

Now that we have gone over how atoms bond to form molecules and how UV-Vis spectroscopy can help characterize electronic transitions, it is time to apply this knowledge directly to our PiB model. Since the “box” needs a defined length for the electrons to move freely in, we can define this length as the length of the proposed molecules. This can help us form a hypothesis of what types of energy we should expect will be absorbed when we run them through the spectrometer. If we then make the simplifying assumption that we can fit all Pi-electrons into our box, we can then figure out the approximate HOMO and LUMO energy levels based of their stationary states using the Pauli exclusion principle. This allows us to figure out the constants needed to define many of the parts of the PiB model. We can then apply the PiB model to the spectroscopic data we collected in the lab to define where the pi -> pi* transition (HOMO to LUMO transition) occurs by looking at the absorption peaks.

The dimensions of the box here are the lengths in x and y of the molecule in question.
In order to decribe the molecular wavefunctions so that we can find the relevent energy levels, we need to  know the length of the box. Electrons in the molecule associated with Pi-bonds do not leave the molecule as it would take energy to break them free that they do not gain without outside inteference. This concept of electrons being stuck in an 2-D region due to a potential energy barrier lends itself well to a PiB description.
To find the length we remember that the electrons are in sp2 orbitals so we have a bond angle between carbons of 120 degrees(see figure below).

Using this as the box lengths x and y we can determine the energy for different n_x and n_y levels and to solve for the stationary states associated with those energies.

## 1-D PiB

```python

print('Enter well length in Å') # See python3 documentation for e notation information
l = float(input())
print('Enter n value')
n = int(input())
x = np.linspace(0,l,50)
psi = np.zeros_like(x)

for i,k in enumerate(x): 
    
    psi[i] = math.sqrt(2/l)*math.sin(n*math.pi*x[i]/l) 
 
 
Energy = ((n**2)*(h**2)/(8*m*(l**2)))*(6.242*10**(-18))
Energy_scino = "{:.2e}".format(Energy) # write in scientific notation with 2 decimals
fig,ax = subplots()
ax.plot(x,psi,label ='ψ(x)') # divide psi and x by 1e-10 to convert them to angstroms
box_height = np.max(psi)*3 # Makes sure that the box is a good height compared to Psi.
plt.vlines(x=0, ymin = -2*np.max(psi), ymax =box_height ,color='k') # Set box height
plt.vlines(x=l, ymin = -2*np.max(psi), ymax = box_height, color ='k')
plt.axhline(y=0 , color='k')
plt.text(0, -np.max(psi), 0,fontsize = 15) # Makes sure that 0 and L labels are visible
plt.text(l,-np.max(psi),'L',fontsize = 15)
plt.legend()
plt.ylim = (-1,2)
text_x = l/3
text_y = (2*box_height/3)
plt.text(text_x,text_y, "E_"+str(n) +" ="+ str(Energy_scino)+"J",fontsize=15) # Show energy on graph
plt.title("Particle in a 1-D box of length L", fontsize=20)
plt.xlabel("angstroms",fontsize=15)
plt.ylabel("ψ(x) " ,fontsize=15)

plt.show()
```

#### 2-D Particle in a Box Model

```python
print('enter l_x in angstroms ')
l_x = float(input())
print('enter l_y in angstroms')
l_y = float(input())
print('Enter nx')
n_x = int(input())
print('enter n_y')
n_y = int(input())

x = np.linspace(0,l_x,1000)
y = np.linspace(0,l_y,1000)

def psi(x,y):
    
    return math.sqrt(2/l_x)*np.sin(n_x*np.pi*x/l_x)*math.sqrt(2/l_y)*np.sin(n_y*np.pi*y/l_y)

X,Y = np.meshgrid(x,y)
Z=psi(X,Y)
Energy_2d = ((hbar**2*np.pi**2)/(2*m))*((n_x**2)/(l_x**2)+(n_y**2)/(l_y**2))*(6.242*10**(-18))
Energy_scino_2d = "{:.2e}".format(Energy_2d)


fig = plt.figure()
ax = plt.axes(projection ='3d')
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5, aspect=5)

print('ΔE =', str(Energy_scino_2d)+" J")
plt.title("Particle in a 2-D box of length L", fontsize=20)
ax.set_ylabel('l_x (Å)', fontsize=15)
ax.set_xlabel("angstroms",fontsize=15)
ax.set_zlabel("ψ(x) " ,fontsize=15)
ax.text(0,l_y,np.max(Z), 'ΔE=' + str(Energy_scino_2d)+" J", fontsize=15)
plt.gcf().set_size_inches(13, 12)
```
